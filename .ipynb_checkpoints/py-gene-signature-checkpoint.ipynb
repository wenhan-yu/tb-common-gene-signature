### 4. Search a criteria for gene set selection

##### 1. Consider all top genes in each comparisons (rule_geneselection1) -- not considered
- Search number of top genes (from top 1 to 100) in up- and down-regulated genes in each comparisons.
- Consider all selected genes [ATB_HC ‚à™ ATB_LTBI ‚à™ ATB_OD ‚à™ ATB_Tret]

##### 2. Consider the top genes not unique to each compaursions (rule_geneselection2) -- not considered
- Search top genes in up- and down-regulated genesets in each comparisons.
- Remove the genes only unique to the individual comparison
 * [(ATB_HC ‚à© ATB_LTBI) ‚à™ (ATB_HC ‚à© ATB_OD) ‚à™ (ATB_HC ‚à© ATB_Tret) ‚à™ (ATB_LTBI ‚à© ATB_OD) ‚à™ (ATB_LTBI ‚à© ATB_Tret) ‚à™ (ATB_OD ‚à™ ATB_Tret)]

##### 3. Consider the top 5% genes in the weighted degree distribution (power-law distribution) and in addition, not unique to each comparsions (rule_geneselection3)
- Assume this covariate network is a scale-free network. A power-law distribution (a generalized beta distribution) is used to model the degree distributionw
- Determine the threshold of 95% of the distribution and find the genes whose are in top 5 % and abs(logFC)>=0.5 in each comparisons 
- Remove the genes only unique to the individual comparison
 * [(ATB_HC ‚à© ATB_LTBI) ‚à™ (ATB_HC ‚à© ATB_OD) ‚à™ (ATB_HC ‚à© ATB_Tret) ‚à™ (ATB_LTBI ‚à© ATB_OD) ‚à™ (ATB_LTBI ‚à© ATB_Tret) ‚à™ (ATB_OD ‚à™ ATB_Tret)]
